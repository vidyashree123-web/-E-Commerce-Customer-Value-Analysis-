```python
# Collapse multiple payment installment rows into ONE total per order
payment_totals = df.groupby('order_id')['payment_value'].sum().reset_index()
payment_totals.columns = ['order_id', 'total_payment_value']

# Now building  a clean ORDER-LEVEL table (one row per order, not per item)
order_level = df.drop_duplicates(subset='order_id')[
    ['order_id','order_status','order_purchase_timestamp',
     'order_delivered_customer_date','customer_unique_id',
     'customer_state','review_score']
].merge(payment_totals, on='order_id', how='left')

print(order_level.shape)
print(order_level['order_id'].duplicated().sum())  # should be 0
```

```python
order_level['order_delivered_customer_date'] = pd.to_datetime(
    order_level['order_delivered_customer_date'], errors='coerce'
)
print(order_level.dtypes)
print("Nulls after conversion:", order_level['order_delivered_customer_date'].isnull().sum())
```

```python
# Reference date = 1 day after the most recent order in the dataset
reference_date = order_level['order_purchase_timestamp'].max() + pd.Timedelta(days=1)
print("Reference date:", reference_date)

rfm = order_level.groupby('customer_unique_id').agg(
    last_purchase=('order_purchase_timestamp', 'max'),
    frequency=('order_id', 'count'),
    monetary=('total_payment_value', 'sum')
).reset_index()

rfm['recency'] = (reference_date - rfm['last_purchase']).dt.days

print(rfm.shape)
print(rfm[['recency','frequency','monetary']].describe())
```

```python
rfm.sort_values('monetary', ascending=False).head(5)
```

```python
# Recency: lower is better, so reverse the labels
rfm['R_score'] = pd.qcut(rfm['recency'], q=5, labels=[5,4,3,2,1]).astype(int)

# Monetary: higher is better, normal quintiles
# .rank(method='first') breaks ties so qcut doesn't fail on duplicate values
rfm['M_score'] = pd.qcut(rfm['monetary'].rank(method='first'), q=5, labels=[1,2,3,4,5]).astype(int)

# Frequency: custom scoring since it's heavily skewed toward 1
def freq_score(f):
    if f == 1:
        return 1
    elif f == 2:
        return 2
    else:
        return 3

rfm['F_score'] = rfm['frequency'].apply(freq_score)

print(rfm[['recency','R_score','frequency','F_score','monetary','M_score']].head(10))
print(rfm['F_score'].value_counts())
```

```python
rfm['RFM_score'] = rfm['R_score'].astype(str) + rfm['F_score'].astype(str) + rfm['M_score'].astype(str)

def segment(row):
    r, f, m = row['R_score'], row['F_score'], row['M_score']
    
    if f >= 2 and r >= 4 and m >= 4:
        return 'Champions'
    elif f >= 2 and r >= 3:
        return 'Loyal Customers'
    elif f >= 2 and r <= 2:
        return 'At Risk (Repeat)'
    elif f == 1 and r >= 4 and m >= 4:
        return 'High-Value New'
    elif f == 1 and r >= 4:
        return 'New Customers'
    elif f == 1 and r <= 2 and m >= 4:
        return 'High-Value Lost'
    elif f == 1 and r <= 2:
        return 'Lost'
    else:
        return 'Average / Needs Attention'

rfm['segment'] = rfm.apply(segment, axis=1)

print(rfm['segment'].value_counts())

seg_revenue = rfm.groupby('segment').agg(
    customers=('customer_unique_id','count'),
    total_revenue=('monetary','sum')
).sort_values('total_revenue', ascending=False)
seg_revenue['revenue_pct'] = (seg_revenue['total_revenue'] / seg_revenue['total_revenue'].sum() * 100).round(1)
seg_revenue['customer_pct'] = (seg_revenue['customers'] / seg_revenue['customers'].sum() * 100).round(1)
print(seg_revenue)
```

```python
import matplotlib.pyplot as plt

# Recompute cleanly from your actual rfm table (run this in your notebook)
seg_summary = rfm.groupby('segment').agg(
    customers=('customer_unique_id','count'),
    total_revenue=('monetary','sum')
).reset_index()
seg_summary['revenue_pct'] = seg_summary['total_revenue'] / seg_summary['total_revenue'].sum() * 100
seg_summary['customer_pct'] = seg_summary['customers'] / seg_summary['customers'].sum() * 100
seg_summary = seg_summary.sort_values('total_revenue', ascending=False)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].barh(seg_summary['segment'], seg_summary['customer_pct'], color='#4C72B0')
axes[0].set_xlabel('% of Customers')
axes[0].set_title('Customer Base: % by Segment')
axes[0].invert_yaxis()

axes[1].barh(seg_summary['segment'], seg_summary['revenue_pct'], color='#DD8452')
axes[1].set_xlabel('% of Total Revenue')
axes[1].set_title('Revenue Contribution: % by Segment')
axes[1].invert_yaxis()

plt.tight_layout()
plt.savefig('rfm_segment_comparison_final.png', dpi=150)
plt.show()
```
